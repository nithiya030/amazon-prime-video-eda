````python
#Box plot to find the outliers to decide whether filling with mean/median for titles dataset

fig,axes=plt.subplots(2,2,figsize=(10,10))
sns.boxplot(ax=axes[0,0],data=titles,x='imdb_score',color='skyblue')
axes[0,0].set_title("Distribution of IMDb Scores")
sns.boxplot(ax=axes[0,1],data=titles,x='imdb_votes',color='skyblue')
axes[0,1].set_title('Distribution of IMDb Votes')
sns.boxplot(ax=axes[1,0], data=titles, x='tmdb_score',color='red')
axes[1,0].set_title('Distribution of TMDB Scores')
sns.boxplot(ax=axes[1, 1], data=titles, x='tmdb_popularity',color='green')
axes[1, 1].set_title('Distribution of TMDB Popularity')
plt.tight_layout()
plt.show()
````

````python
# Impute rating scores and popularity metrics using their median values due to outliers
titles['imdb_score'] = titles['imdb_score'].fillna(titles['imdb_score'].median())
titles['tmdb_score'] = titles['tmdb_score'].fillna(titles['tmdb_score'].median())
titles['imdb_votes'] = titles['imdb_votes'].fillna(titles['imdb_votes'].median())
titles['tmdb_popularity'] = titles['tmdb_popularity'].fillna(titles['tmdb_popularity'].median())
````

````python
# Impute Age Certification column with 'Unrated' value
titles['age_certification']=titles['age_certification'].fillna('Unrated')
````

````python
# Drop Description and IMDb ID as they are not useful for further analysis

titles.drop(columns=['description', 'imdb_id'], inplace=True)
````

````python
# Calculating the mean percentage of missing values of titles dataset
titles.isnull().mean()*100
````

````python
# impute character column with 'Unknown' if role is character because director doesn't have roles
actor_mask = credits['role'] == 'ACTOR'
credits.loc[actor_mask, 'character'] = credits.loc[actor_mask, 'character'].fillna('Unknown')
````

### What all manipulations have you done and insights you found?

Duplicates from both datasets have been removed.

Created box plots to identify outliers and determine whether to fill missing values using the mean or median in the titles dataset. Since all four continuous data columns contained extreme outliers, the median was imputed for the null values. For age_certification, which had 65% missing values, a placeholder value of 'Unrated' was used instead of dropping the column so that maturity ratings can be analyzed against scores later.

Dropped the description and imdb_id columns because they do not provide statistical value for further analysis. Imputed the character column in the credits dataset with 'Unknown' only when the role is 'ACTOR', as directors do not have character roles.

The genres column was originally stored as a string data type, so it was converted into a true Python list of strings. A new DataFrame, genre_analysis_df, was created to explode the genres. This step is necessary because exact genre counts cannot be calculated while they are locked inside lists. In this new DataFrame, .str.strip() was applied to eliminate accidental leading or trailing whitespaces, and the resulting null values were dropped since they are unnecessary for genre-specific analysis.

Similarly, the production_countries column was converted from a string to a list of strings and exploded into a dedicated country_analysis_df to accurately calculate country-level counts. The .str.strip() function was used to clear whitespaces, and empty rows were dropped.

Finally, both datasets were combined into a unified DataFrame named merged_df using an inner join on the id column to prepare for comprehensive actor, director, and content mapping.
